# Initialisation 
Cette partie du notebook concerne :
* le dossier de travail a renseigner (dossier ou se trouve les fichiers outlines.txt)
* les valeurs a remplacer
* la ligne concernee 

```python
#ligne du document ouline à modifier
target_line = 4 # en python la ligne 0 correspond a la ligne 1 du fichier .txt 

#valeurs à trouver et à changer dans le fichier outline
old_val = "C3.14_ON_01_w1SPI-488_s4.C3.14_ON_01_w1SPI-488_s4_tif" 
new_val = ".tif"

#chemin du dossier _FQ_outline
chemin = "/media/raphael/data/works/20231220_C3_14_2h_ofON/bleach_correction/expo/Hyperstack_splitter_folder" + "/"


#vérification de la présence des fichiers outline dans le dossier
list = os.listdir(path=chemin)
list
```

# Remplacement des noms et des lignes 


## Remplacement des lignes 
Cette partie du notebook correspond aux code qui va permettre de remplacer dans chaque fichiers outline.txt l'element old_val par new_val a la ligne correspondant a target_line

```python
#Fonction qui parcours le dossier en changeant les valeurs definies au début du script
for i in list:
    with open(chemin+i, "r", encoding = "utf-8") as fichier:
        contenu = fichier.read()
        lignes_contenu = contenu.split("\n")
        lignes_contenu[target_line] = lignes_contenu[target_line].replace(old_val,new_val)
        contenu = "\n".join(lignes_contenu)
        with open(chemin+i, "w", encoding = "utf-8") as fichier:
            fichier.write(contenu)
            print("Done for " +i)
```

## Remplacement des noms 
Cette partie du notebook permet de remplacer tout les noms de fichiers present dans un repertoire d'une maniere similaire a celle precedement enoncee en remplacant old_val par new_val.

Pour cela on utilise os.chdir() pour changer de repertoire de travail et se situer la ou se situent les fichiers outline.txt

```python
os.chdir(chemin) 
os.getcwd()
```

```python
def rename_file(name_file,new_name):
    # change definitivement le nom du fichier
    os.rename(name_file,new_name)
    print("Done for " + new_name)

def replace_pattern_file(pattern, replace,name_file) : 
    # Remplace le motif old_val par new_val uniquement dans la liste de string list a un indice donnee
    new_name = re.sub(pattern, replace, name_file)
    print('ancien nom : ', name_file, '\nnouveau nom : ', new_name)
    return new_name

pattern = old_val #[0:-4] # correspond a 'w3SPI-488.tif' moins les 4 derniers caracteres soit 'w3SPI-488' 
replace = new_val #[0:-4] # correspond a 'w1SPI-642.tif' moins les 4 derniers caracteres soit 'w1SPI-642' 

for i in list : 
    
    new_name = replace_pattern_file(pattern,replace,i)
    rename_file(i,new_name)
    
    
```
